```python
import tensorflow as tf 
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

#simplest possible neural network - one layer, one neuron
#sequential = defines layers, inside sequential - define what each layer look like
#using the keras - define what the layer look like, dense means a set of fully densely connected
#neurons - every neuron is connected to other neuron 
model = Sequential([Dense(units=1, input_shape=[1])])
#how good the quess is - loss function
#sgd = optimizer - stochastic gradient descent - complex math function that, when given the values
# the previous guess, and the result of calculating the errors on that guess, can then generate another one
# the job of the optimizer is to minimize the loss -> bring guessed formula closer and closer
model.compile(optimizer='sgd', loss='mean_squared_error')

#input data, format that the layers expect -> numpy array
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

#fit the Xs to the Ys and try it 500 times
model.fit(xs, ys, epochs=500)
#predict Y when X is 10 - value close to 19
print(model.predict([10.0]))
```

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

#variale called l0 to hold dense layer, after the network finishes learning, I can print
#out the values or weight I learned
l0 = Dense(units=1, input_shape=[1])
model = Sequential([l0])
model.compile(optimizer = 'sgd', loss = 'mean_squared_error')

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
xy = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

model.fit(xs, xy, epochs = 500)

print(model.predict([10.0]))
print('Here is what I learned: {}'.format(l0.get_weights()))
```
